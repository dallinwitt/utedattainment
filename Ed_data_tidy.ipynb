# Data Tidying
The purpose of this code is to restructure the dataset to compley with "tidy" data standards, namely that each row represents a single observation, and each column represents a metric of that observation. This dataset was pretty far from tidy, but was fixable with relatively few lines of code.

```python
import pandas as pd
```

```python
ed = pd.read_csv('Utah_Educational_Level_By_Census_Tract_2013_multi_locconcat.csv', header = None)
fips = pd.read_csv('utah_fips.csv')
```

```python
ed
```

```python
ed_T = ed.transpose()
```

```python
ed_T.columns = ed_T.iloc[0,:]
ed_T.drop(0, axis = 0, inplace = True)
ed_T
```

```python
ed_multi = ed_T.set_index(['sex', 'age', 'metric'])
```

```python
ed_multi
```

```python
ed_multi_stack = ed_multi.stack()
ed_multi_stack
```

```python
ed_multi_stack.to_csv('ed_tidy.csv')
```
